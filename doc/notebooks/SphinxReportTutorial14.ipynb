We have copied the snipped from the sphinxreport-test command below:

```python
%matplotlib inline
import SphinxReport.test
result = SphinxReport.test.main( do_print = False,
tracker="TrackerDataframes",
renderer="none",
trackerdir="/ifs/devel/andreas/sphinx-report/doc/trackers",
workdir="/ifs/devel/andreas/sphinx-report/doc",
glob="data/*.tsv.gz", regex="data/(.*).transcriptprofile*")
```

The data is now stored in a pandas dataframe called result.

```python
print result.head()
```

Plotting from dataframe is very conveniently done with the ggplot package. First, we neet to import ggplot.

```python
from ggplot import *
```

We can now try a simple graph.

```python
ggplot( aes( x='bin', y='none'), data=result ) + geom_point()
```

Our first plot is not very informative - we can do better. Let us colour the data points by experiremnt or 'track'. Currently,
this information is part of the hierarchical index of the dataframe. For ggplot to be able to make use of that information we need to add this information into an additional column. The pandas.Dataframe.reset_index method does exactly that:

```python
result.reset_index( 0, inplace = True )
result.columns
```

Note how there is now a new column called 'track'. We can use the track information to annotate the plot.

```python
ggplot( aes( x='bin', y='none', color='track'), data=result ) + geom_point()
```

Much better. We notice that the profiles are not normalized for read-depth and are thus not directly comparable. So let us try plotting the normalized profiles. First, we plot profiles normalized by area.

```python
ggplot( aes( x='bin', y='area', color='track'), data=result ) + geom_point()
```

We can see that the background levels still differ between profiles. What has happenden is that even profiles will be close to an average, while profiles with peaks will depress the background. Let us try background normalization instead, which uses the profile densities at the extreme ends for normalization.

```python
ggplot( aes( x='bin', y='background', color='track'), data=result ) + geom_point()
```

This normalization allows us to compare profiles between experiments with different read dephts. We now want to put the plot above into our sphinx-report report. We can use the ggplot renderer. Note that this renderer automatically adds the track information as a column. Thus, we only need to take note of the parameters for the plot, the aesthatics and the geometry

* aes: x='bin', y='background', color='track'
* geom: geom_point()
