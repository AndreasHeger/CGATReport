# Further investigation of methylation profiles

Connecting to the CGATReport data source:

```python
%matplotlib inline
import CGATReport.test
result = CGATReport.test.main( do_print = False,
tracker="TrackerDataframes",
renderer="none",
trackerdir="/ifs/devel/andreas/sphinx-report/doc/trackers",
workdir="/ifs/devel/andreas/sphinx-report/doc",
glob="data/*.tsv.gz", regex="data/(.*).transcriptprofile*")
```

The data is now stored in a pandas dataframe called result.

```python
print result.head()
```

Plotting from dataframe is very conveniently done with the ggplot package. First, we neet to import ggplot.

```python
from ggplot import *
```

For ggplot to be able to make use of the track information, we need to add the index as an additional column. The *pandas.Dataframe.reset_index* method does exactly that:

```python
result.reset_index( 0, inplace = True )
print result.head()
```

Note how there is now a new column called 'track'. We can use the track information to annotate the plot.

```python
ggplot( aes( x='bin', y='none', color='track'), data=result ) + geom_point()
```

These are the unnormalized profiles and much less comparable than the background normalized profiles.
